```python
import asyncio

async def fetch_data(delay):
    print("Fetching Data...")
    await asyncio.sleep(delay)
    print("Data Fetched!")
    return {"data": "some data"}


async def main():
    print("Start of main coroutine")
    task = fetch_data(2)
    result = await task
    print(f"Received result: {result}")
    print("End of main coroutine")


asyncio.run(main())
```
